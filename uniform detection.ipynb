```python
# Import necessary libraries
import cv2
from ultralytics import YOLO
from google.colab.patches import cv2_imshow

# Load the trained YOLO model
model = YOLO("/content/runs/detect/train/weights/best.pt")

# Start capturing video from the webcam (use `0` for the default webcam)
cap = cv2.VideoCapture(0)

# Check if the camera is opened successfully
if not cap.isOpened():
    print("Error: Could not open webcam.")
else:
    while True:
        # Capture each frame from the webcam
        ret, frame = cap.read()

        if not ret:
            print("Error: Failed to capture image.")
            break

        # Perform object detection on the frame
        results = model(frame)  # Using the YOLO model for inference

        # Display the results (draw boxes, labels, etc.)
        results.render()  # This will draw bounding boxes on the frame
        frame = results.imgs[0]  # Get the frame with detections

        # Display the frame with detections
        cv2_imshow(frame)  # Show the frame in Colab notebook
        cv2.waitKey(1)  # Wait for 1 ms to process keypress (you can use ESC to stop)

        # Exit the loop if 'q' is pressed (in Colab you can't directly use OpenCV's cv2.waitKey)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

# Release the video capture and close the display window
cap.release()
cv2.destroyAllWindows()
```
